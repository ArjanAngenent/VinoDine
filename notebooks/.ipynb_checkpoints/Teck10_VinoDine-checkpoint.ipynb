```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
csv_file_path = '~/code/ArjanAngenent/VinoDine/raw_data/XWines_Full_100K_wines.csv'
wine_df = pd.read_csv(csv_file_path)
wine_df['Harmonize'] = wine_df[['Harmonize']].apply(lambda string: string.str.strip("[").str.rstrip("]").str.replace("'","").str.replace(" ",""))
wine_df['Harmonize'] = wine_df['Harmonize'].str.split(",")
wine_df['Grapes'] = wine_df[['Grapes']].apply(lambda string: string.str.strip("[").str.rstrip("]").str.replace("'","").str.replace(" ",""))
wine_df['Grapes'] = wine_df['Grapes'].str.split(",")
wine_df.head()
```

```python
plots = ['Type', 'Body', 'Acidity', 'Country', 'RegionName']
fig, axs = plt.subplots(len(plots), figsize=(10,50))
i=0
for plot in plots:
    plt.title(plot)
    axs[i].hist(wine_df[plot])
    plt.show;
    i += 1
```

# 1. Create binary categories for Harmonize

```python
harmonize_df = wine_df['Harmonize']

food_cats = []
for row in harmonize_df:
    for cat in row:
        cat = cat.strip()
        if cat not in food_cats:
            food_cats.append(cat)

food_cats, len(food_cats)
```

```python
for ind, row in wine_df.iterrows():
    wine_df.loc[ind,row['Harmonize']] = 1

wine_df[food_cats] = wine_df[food_cats].fillna(0).astype(np.int16)

wine_df.head()
```

# 2. Create binary categories for Grapes

```python
grapes_df = wine_df['Grapes']

grape_cats = []
for row in grapes_df:
    for cat in row:
        cat = cat.strip()
        if cat not in grape_cats:
            grape_cats.append(cat)

grape_cats, len(grape_cats)
```

```python
for ind, row in wine_df.iterrows():
    wine_df.loc[ind,row['Grapes']] = 1

wine_df[grape_cats] = wine_df[grape_cats].fillna(0).astype(np.int16)

wine_df.head()
```

```python
save_file_name = "Cleaned_Full_100K_wines"
wine_df.to_csv(f'~/code/ArjanAngenent/VinoDine/raw_data/{save_file_name}.csv', index=False)
```
