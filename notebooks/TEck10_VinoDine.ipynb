```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
csv_file_path = '~/code/ArjanAngenent/VinoDine/raw_data/XWines_Full_100K_wines.csv'
wine_df = pd.read_csv(csv_file_path)
wine_df['Harmonize'] = wine_df[['Harmonize']].apply(lambda string: string.str.strip("[").str.rstrip("]").str.replace("'","").str.replace(" ",""))
wine_df['Harmonize'] = wine_df['Harmonize'].str.split(",")
wine_df['Grapes'] = wine_df[['Grapes']].apply(lambda string: string.str.strip("[").str.rstrip("]").str.replace("'","").str.replace(" ",""))
wine_df['Grapes'] = wine_df['Grapes'].str.split(",")
wine_df
```

```python
save_file_name = "Cleaned_Full_100K_wines"
wine_df.to_csv(f'~/code/ArjanAngenent/VinoDine/raw_data/{save_file_name}.csv', index=False)

clean_wine_df = pd.read_csv(f'~/code/ArjanAngenent/VinoDine/raw_data/{save_file_name}.csv')

clean_wine_df
```

# Plotting

```python
plots = ['Type', 'Body', 'Acidity', 'Country', 'RegionName']
fig, axs = plt.subplots(len(plots), figsize=(10,50))
i=0
for plot in plots:
    plt.title(plot)
    axs[i].hist(wine_df[plot])
    plt.show;
    i += 1
```

# exploded file

```python
wine_df_exp = wine_df.explode('Harmonize')

wine_df_exp = wine_df_exp.explode('Grapes')

harm_values= wine_df_exp['Harmonize'].value_counts()

selected_harmonize_values = harm_values[harm_values>=500].index.to_list()

grape_values= wine_df_exp['Grapes'].value_counts()

selected_grape_values = grape_values[grape_values>=2_000].index.to_list()

wine_df_exp = wine_df_exp.loc[(wine_df_exp['Harmonize'].isin(selected_harmonize_values))]

wine_df_exp = wine_df_exp.loc[(wine_df_exp['Grapes'].isin(selected_grape_values))]
wine_df_exp.head()
```

```python
plt.figure(figsize=(50,15))
plt.hist(wine_df_exp['Harmonize'])
plt.ylim(bottom=20_000);
```

```python
save_file_name = "Cleaned_Full_100K_wines_exploded"
wine_df_exp.to_csv(f'~/code/ArjanAngenent/VinoDine/raw_data/{save_file_name}.csv', index=False)
```
